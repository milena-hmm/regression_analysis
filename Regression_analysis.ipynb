Создайте новый признак – марку автомобиля (company). Машины каких производителей встречаются в датасете? Далее исправьте названия и проверьте изменения.

```python
cars['company'] = cars.CarName.apply(lambda x: x.split(' '))
```

```python
cars.tail()
```

```python
cars['company'] = cars.company.apply(lambda x: x[0])
```

```python
cars.company.nunique()
```

```python
cars = cars.drop(['CarName', 'car_ID'], axis=1)
```

```python
cars.company.unique()
```

```python
cars = cars.replace('maxda', 'mazda').replace('Nissan', 'nissan'). \
    replace('porcshce', 'porsche').replace('toyouta', 'toyota').replace(['vokswagen', 'vw'], 'volkswagen')
```

```python
cars.company.nunique()
```

```python
cars.head(1)
```

Оставьте только часть предикторов, после чего посчитайте корреляцию между price и другими переменными.

```python
cars = cars.drop(['symboling','doornumber','enginelocation','carheight', 'fuelsystem', 'stroke', 'compressionratio', 'peakrpm', 'citympg', 'highwaympg'], axis=1)
```

```python
cars.corr()
```

```python
cars.dtypes
```

Преобразуйте категориальные переменные с помощью pd.get_dummies().

```python
df_dummy = pd.get_dummies(data=cars[['fueltype', 'aspiration', 'carbody' , 'drivewheel', 'enginetype', 'cylindernumber', 'company']], drop_first = True)
```

```python
df_dummy
```

```python
cars_dummy = pd.concat([cars.drop(['fueltype', 'aspiration', 'carbody' , 'drivewheel', 'enginetype', 'cylindernumber', 'company'], axis=1), df_dummy], axis=1)
```

```python
cars_dummy.shape
```

Постройте модель с одним предиктором цены – horsepower. Какой процент изменчивости объясняет полученная модель? (\(R^2\))

```python
results = smf.ols('price ~ horsepower', data=cars_dummy).fit()

print(results.summary())
```

Далее – две модели (со всеми предикторами и со всеми, кроме марок машин). Обратите внимание на изменения в \(R^2\), коэффициентах и их значимости.

```python
sm.add_constant(cars_dummy)
results = smf.ols('price ~ wheelbase + carlength + carwidth + curbweight + enginesize + boreratio + horsepower + fueltype_gas + aspiration_turbo + carbody_hardtop + carbody_hatchback + carbody_sedan + carbody_wagon + drivewheel_fwd + drivewheel_rwd + enginetype_dohcv + enginetype_l + enginetype_ohc + enginetype_ohcf + enginetype_ohcv + enginetype_rotor + cylindernumber_five + cylindernumber_four + cylindernumber_six + cylindernumber_three + cylindernumber_twelve + cylindernumber_two + company_audi + company_bmw + company_buick + company_chevrolet + company_dodge + company_honda + company_isuzu + company_jaguar + company_mazda + company_mercury + company_mitsubishi + company_nissan + company_peugeot + company_plymouth + company_porsche + company_renault + company_saab + company_subaru + company_toyota + company_volkswagen + company_volvo', cars_dummy).fit()
print(results.summary())
```

```python
sm.add_constant(cars_dummy)
results = smf.ols('price ~ wheelbase + carlength + carwidth + curbweight + enginesize + boreratio + horsepower + fueltype_gas + aspiration_turbo + carbody_hardtop + carbody_hatchback + carbody_sedan + carbody_wagon + drivewheel_fwd + drivewheel_rwd + enginetype_dohcv + enginetype_l + enginetype_ohc + enginetype_ohcf + enginetype_ohcv + enginetype_rotor + cylindernumber_five + cylindernumber_four + cylindernumber_six + cylindernumber_three + cylindernumber_twelve + cylindernumber_two', cars_dummy).fit()
print(results.summary())
```
